```python
yearly_crimes = df.groupby('Year')['Total_Crimes'].sum().reset_index()
sns.barplot(x='Year', y='Total_Crimes', data=yearly_crimes)
plt.title('Total Crimes in India by Year')
plt.ylabel('Total Crimes')
plt.show()
```

```python
state_crimes = df.groupby('State/UT')['Total_Crimes'].sum().sort_values(ascending=False)
state_crimes.plot(kind='bar', figsize=(10,6))
plt.title('Total Crimes by State')
plt.ylabel('Total Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
women_crimes = df.groupby(['Year'])['Crimes_Against_Women'].sum().reset_index()
sns.lineplot(x='Year', y='Crimes_Against_Women', data=women_crimes, marker='o')
plt.title('Crimes Against Women Over Time')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
corr = df[['Total_Crimes', 'Crimes_Against_Women', 'Murders', 'Theft', 'Cyber_Crimes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Crime Type Correlation Matrix')
plt.show()
```

```python
reshaped_df = df.pivot_table(
    index="Murders", 
    columns="Year", 
    values="Total_Crimes", 
    aggfunc='sum'  # or 'mean', 'max', etc.
)
reshaped_df.head()
```

```python
reshaped_df.fillna(0,inplace=True)
```

```python
 import matplotlib.pyplot as plt
```

```python
plt.plot(reshaped_df.index,reshaped_df["2019"])
```

```python
plt.plot(reshaped_df.index, reshaped_df[2019])
plt.title('Total Crimes vs Murders in 2019')
plt.xlabel('Murders')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(reshaped_df.index, reshaped_df[2020], marker='o', linestyle='-')
plt.title("Total Crimes vs Murders in 2020", fontsize=18)
plt.xlabel("Murders", fontsize=16)
plt.ylabel("Total Crimes", fontsize=16)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Murders", fontsize=14)
plt.ylabel("Total Crimes", fontsize=14)
plt.ylim(0, 350000) 


plt.plot(reshaped_df.index, reshaped_df[2020], marker='o', color='blue', label='2020')
plt.title("Total Crimes vs Murders in 2020", fontsize=18)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Murders", fontsize=14)
plt.ylabel("Total Crimes", fontsize=14)
plt.ylim(0, 350000)


plt.plot(reshaped_df.index, reshaped_df[2020], label='Year 2020', marker='o')
plt.plot(reshaped_df.index, reshaped_df[2021], label='Year 2021', marker='o')

plt.title("Total Crimes vs Murders in 2020 and 2021", fontsize=18)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Murders", fontsize=14)
plt.ylabel("Total Crimes", fontsize=14)
plt.ylim(0, 350000) 

for column in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[column], label=str(column), linewidth=2)

plt.title("Total Crimes vs Murders (All Years)", fontsize=18)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
